## Filtering and ordering

```python
df.rename(columns = {' "Year"': 'year', ' "Age"': 'age', ' "Name"': 'name'}, inplace = True)

df[['year']] = df[['year']].apply(pd.to_numeric)

## filter by age >= 60

df[df['year'] >= 60]

```

```python
## lookink for certain data filtering
specific_ages = [30,40]
df[df['year'].isin(specific_ages)]
```

```python
## setting an index
df2 = df.set_index('name')

## filtering by cloumns
df2.filter(items = ['year'])
```

```python
## order by
df[df['year'] >= 55].sort_values(by = 'year', ascending = False)
```
