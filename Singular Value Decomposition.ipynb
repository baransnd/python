## Compression of a picture through numerical Simgular-Value-Decomposition
The picture will be represented by a matrix A representing each pixel with each cell,
then A decomposed into 3 matrices U, S, Vh; where S is the singular-value matrix,
which allows the picture to be compressed through compression of S.

```python
import datetime

import numpy as np
import matplotlib.pyplot as plt
from scipy import datasets
```

getting a random picture from scipy

```python
image = datasets.ascent()
u, s, vh = np.linalg.svd(image)
image_size = image.size
```

Below is the method to calculate how much compressed matrix_ak is in comparrision to the original matrix A.

```python
def frobenius(matrix_a, matrix_ak):
    rows = len(matrix_a)
    columns = len(matrix_a[0])
    sum = 0
    for i in range(rows):
        for j in range(columns):
            sum += np.absolute(matrix_a[i][j] - matrix_ak[i][j])**2
    sum = np.sqrt(sum)
    return sum
```

Below is the calculating method for compressed images, given how many singular values in the compressed images is to be used

```python
def calculate_compressed_images(matrix_sizes):
    for k in matrix_sizes:
        # Create the rank-k approximation matrix
        diagonal_matrix = np.diag(s[:k])
        k_compressed_image = u[:, :k] @ diagonal_matrix @ vh[:k, :]

        # calculate the size of the compressed image
        compressed_size = (u[:, :k].size + diagonal_matrix.size + vh[:k, :].size)

        #show the compressed image
        plt.imshow(k_compressed_image)
        plt.title('R = ' + str(k))
        plt.axis('off')
        plt.gcf().text(.4, .05, "Compression rate : " + str(compressed_size / image_size))
        plt.gcf().text(.4, .025, "Approximation error: " + str(frobenius(image, k_compressed_image)))
        plt.show()
```

Example numbers of singular values for the compression

```python
matrix_sizes = [5,20,75]
calculate_compressed_images(matrix_sizes)
```
